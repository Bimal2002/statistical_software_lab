```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.stats import chi2
def pdf(x):
    return 20 * x * (1 - x)**3

def beta_function(sample_size):
    samples = []
    i = 0
    while i < sample_size:
        U1 = random.uniform(0, 1)
        U2 = random.uniform(0, 1)
        if U2 <= (256/27) * U1 * (1 - U1)**3:
            samples.append(U1)
            i += 1
    return samples

def simson_one_third(f, n, a, b):
    if n % 2 == 1:
        n += 1
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    integral = h / 3 * (y[0] + y[-1] + 4 * sum(y[1:-1:2]) + 2 * sum(y[2:-1:2]))
    return integral

def beta_cdf(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    else:
        return simson_one_third(pdf, 1000, 0, x)

bins = 50
x_values = np.linspace(0, 1, bins)
cdf_values = [beta_cdf(x) for x in x_values]
# print("CDF values:", cdf_values)

sample_size = 1000
samples = beta_function(sample_size)

breaks = x_values
first_counts, _ = np.histogram(samples, bins=breaks)
first_count=first_counts/sample_size
plt.hist(samples, bins=breaks, edgecolor='black', alpha=0.7, label="First Method")
plt.legend()
plt.show()
# print("First count (histogram counts):", first_count)

expected = []
for i in range(len(breaks) - 1):
    lower_index = np.digitize(breaks[i], x_values) - 1
    upper_index = np.digitize(breaks[i+1], x_values) - 1
    expected_value = cdf_values[upper_index] - cdf_values[lower_index]
    expected.append(expected_value)

# print("Expected values:", expected)

# Chi-square test
def chi_square_test(observed, expected, degrees_of_freedom):
    observed = np.array(observed, dtype=float)
    expected = np.array(expected, dtype=float)

    valid_indices = expected > 0

    chi_squared_statistic = np.sum(
        ((observed[valid_indices] - expected[valid_indices])**2) / expected[valid_indices]
    )

    p_value = 1 - chi2.cdf(chi_squared_statistic, degrees_of_freedom)

    return chi_squared_statistic, p_value

df = len(expected) - 1

chi_first, p_first = chi_square_test(first_count, expected, df)
print("chi_squared_statistics",chi_first)
print("p_value",p_first)

```

```python
from scipy.stats import chisquare, uniform, beta, expon, norm
support = np.arange(1, 10 + 1)
prob = np.array([0.11, 0.12, 0.09, 0.08, 0.12, 0.10, 0.09, 0.09, 0.10, 0.10])
assert len(support) == len(prob)
x = []
NUM_SAMPLES=1000
for _ in range(NUM_SAMPLES):
  while True:
    u1 = uniform.rvs(0, 1)
    y = int(10 * u1) + 1
    u2 = uniform.rvs(0, 1)
    if u2 <= prob[y - 1] / 0.12:
      x.append(y)
      break

x = np.array(x)
x[ : 10]
# print(x)
# print(support)
empirical_prob = np.zeros_like(prob)
for i in range(len(support)):
    empirical_prob[i] = np.sum(x == support[i]) / NUM_SAMPLES

plt.figure(figsize=(8, 6))
plt.plot(support, prob, color='pink', marker='o', label='Original PMF', linestyle='-', markersize=8)
plt.plot(support, empirical_prob, color='red', marker='o', label='Empirical PMF (Generated Samples)', linestyle='-', markersize=8)

plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Line Graph of Original and Empirical PMF')
plt.legend()
plt.grid(True)
plt.show()
observed_frequencies = np.array([np.sum(x == val) for val in support])
expected_frequencies = NUM_SAMPLES * prob
chi2_statistic, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)

print(f"Chi-squared statistic: {chi2_statistic}")
print(f"P-value: {p_value}")
```
